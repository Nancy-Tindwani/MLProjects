# Checking for Outliers

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
sns.boxplot(data["Sales"])
plt.subplot(1,2,2)
sns.boxplot(data["Profit"])
```

# Removing Outliers

```python
for x in data["Sales"]:
    if(x>6000):
        data["Sales"]=data["Sales"].replace(to_replace=x,value=data["Sales"].mean())

sns.boxplot(data["Sales"])
```

```python
for x in data["Profit"]:
    if(x>3000 or x<-2500):
        data["Profit"]=data["Profit"].replace(to_replace=x,value=data["Profit"].mean())

sns.boxplot(data["Profit"])
```

```python
sns.boxplot(data["Discount"])
```

```python
for x in data["Discount"]:
    if(x>0.5):
        data["Discount"]=data["Discount"].replace(to_replace=x,value=data["Discount"].mean())
sns.boxplot(data["Discount"])
```
